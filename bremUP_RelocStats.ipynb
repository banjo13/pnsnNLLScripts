```python
ot = df['OT'].values
loc = df['NLLOC'].values
rms = df['RMS'].values
nphs = df['Nphs'].values
gap = df['Gap'].values
d2 = df['Dist'].values
eLL1 = df['semiMajorAxisLength'].values
eLL2 = df['semiIntermediateAxisLength'].values
eLL3 = df['semiMinorAxisLength'].values
mlat = df['mLat'].values
mlon = df['mLong'].values
mdep = df['mDepth'].values
exlat = df['ExpectLat'].values
exlon = df['exLong'].values
exdep = df['exDepth'].values
evids = df['evids'].values
```

```python
avgLen = []
DIFF = []
maxLen = []
for i in range(len(gap)):
    ellipse = np.array([eLL1[i],eLL2[i],eLL3[i]])
    avg_lenELL = np.average(ellipse)
    mlen = np.max(ellipse)
    avgLen.append(avg_lenELL)
    diff = obspy.geodetics.base.gps2dist_azimuth(mlat[i], mlon[i], exlat[i], exlon[i], a=6378137.0, f=0.0033528106647474805)
    DIFF.append(diff)
    maxLen.append(mlen)
```

```python
dif = []
for i in range(len(DIFF)):
    diff = DIFF[i][0]
    diff = diff/1000
    dif.append(diff)
```

```python
df2 = pd.DataFrame(list(zip(ot,loc,rms,nphs,gap,d2,eLL1,eLL2,eLL3,mlat,mlon,mdep,exlat,exlon ,exdep, evids)),
                      columns =[['ot','loc','rms','nphs','gap','d2ns','eLL1','eLL2','eLL3','mlat','mlon','mdep','exlat','exlon ','exdep','evids']] )
```

```python
# df2.loc[df2['evids'].values == ]
```

```python
df2['DIFF'] = dif
df2['avgLen'] = avgLen
df2['maxLen'] = maxLen
```

```python
df2.to_csv(ofile, index=0)
```
